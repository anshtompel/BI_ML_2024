```python
for i in range(2,len(data.columns)):
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(5, 5), sharex=True, gridspec_kw={"height_ratios": (1, 2)})
    sns.histplot(data.iloc[:, i], ax=ax1)
    ax1.set_title('Histogram')
    sns.boxplot(data.iloc[:, i], ax=ax2, orient='h')
    ax2.set_title('Boxplot')
    plt.show()
```
